# **MODE**

```python
def hitung_total_char(teks):
    total_char = len(teks)
    return total_char

teks_input = "Ini adalah contoh teks untuk dihitung jumlah karakternya."
total_char = hitung_total_char(teks_input)

print(f"Total karakter dalam teks: {total_char}")
```

```python
def hitung_total_kata(teks):
    # Menghapus karakter non-alphanumeric dan merubah seluruh huruf menjadi lowercase
    teks = ''.join(e.lower() if e.isalnum() or e.isspace() else ' ' for e in teks)

    # Menghitung total kata menggunakan split()
    total_kata = len(teks.split())

    return total_kata

# Contoh penggunaan
teks_input = "Ini adalah contoh teks untuk menghitung total kata dalam statistika deskriptif."
total_kata = hitung_total_kata(teks_input)

print(f"Total kata dalam teks: {total_kata}")
```
